```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "stage": ["Linear Regression", "Random Forest"],
    "model": ["LR", "RF"],
    "MAE": [3.3091, 6.3356]
})

sns.set_theme(style="whitegrid")

# consistent model colours (same as previous plot)
palette = {
    "LR": "#F3B089",   # orange
    "RF": "#4C72B0"    # blue
}

plt.figure(figsize=(7, 7))
ax = sns.barplot(
    data=df,
    x="stage",
    y="MAE",
    hue="model",
    palette=palette
)

ax.set_title("Comparing MAE of Initial Linear Regression vs Random Forest Models")
ax.set_ylabel("MAE (cents)")
ax.set_xlabel("")

for container in ax.containers:
    ax.bar_label(container, fmt="%.3f", padding=3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "stage": ["Before Tuning", "Before Tuning", "After Tuning", "After Tuning"],
    "model": ["Linear Regression", "Random Forest", "Linear Regression", "Random Forest"],
    "MAE": [3.3091, 6.3356, 3.3091, 3.2363]
})

sns.set_theme(style="whitegrid")

# fixed colors per model
palette = {
    "Linear Regression": "#F3B089",   # orange
    "Random Forest": "#4C72B0"    # blue (same as before)
}

plt.figure(figsize=(7, 7))
ax = sns.barplot(
    data=df,
    x="stage",
    y="MAE",
    hue="model",
    palette=palette
)

ax.set_title("Comparing MAE of Models Before vs. After Hyperparameter Tuning")
ax.set_ylabel("MAE (cents)")
ax.set_xlabel("")

# value labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.3f", padding=3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "stage": ["Before Tuning", "After Tuning"],
    "MAE": [6.3356, 3.2363]
})

sns.set_theme(style="whitegrid")

plt.figure(figsize=(7, 7))
ax = sns.barplot(
    data=df,
    x="stage",
    y="MAE",
    color="#4C72B0"  # single consistent color
)

ax.set_title("Comparing MAE of Random Forest Model Before vs. After Hyperparameter Tuning")
ax.set_ylabel("MAE (cents)")
ax.set_xlabel("")

ax.bar_label(ax.containers[0], fmt="%.3f", padding=3)

plt.tight_layout()
plt.show()
```
